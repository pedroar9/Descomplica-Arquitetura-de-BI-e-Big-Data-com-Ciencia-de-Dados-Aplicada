```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
amostra['Texto da Avaliação'] = amostra['Texto da Avaliação'].fillna('').astype(str)
```

```python
texto_concatenado = ' '.join(amostra['Texto da Avaliação'])
```

```python
palavras = texto_concatenado.lower().split()
```

```python
frequencias = Counter(palavras)
```

```python
df_frequencias = pd.DataFrame(frequencias.items(), columns=['palavra', 'frequencia'])
```

```python
import spacy
```

```python
nlp = spacy.load('pt_core_news_sm')
```

```python
def encontrar_adjetivos(texto):
    doc = nlp(texto)
    adjetivos = [token.text.lower() for token in doc if token.pos_ == 'ADJ']
    return adjetivos
```

```python
todos_adjetivos = []
for avaliacao in amostra['Texto da Avaliação'].dropna():
    todos_adjetivos.extend(encontrar_adjetivos(avaliacao))
```

```python
frequencias_adjetivos = Counter(todos_adjetivos)
```

```python
df_adjetivos = pd.DataFrame(frequencias_adjetivos.items(), columns=['Adjetivo', 'Frequencia'])
```

```python
df_adjetivos.sort_values(by='Frequencia', ascending=False, inplace=True)
```

```python
df_adjetivos.reset_index(drop=True, inplace=True)
```

```python
df_adjetivos
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(df_adjetivos.set_index('Adjetivo')['Frequencia'])
```

```python

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') 
plt.show()
```
