```python
from elasticsearch import Elasticsearch, helpers
import pandas as pd
import os 
```

```python
es = Elasticsearch(
    "http://localhost:9200",  
    http_auth=('elastic', 'elastic')
)
```

```python
parquet_path = './'
```

```python
def index_parquet_files():
    for filename in os.listdir(parquet_path):
        if filename.endswith('.parquet'):
            # Ler arquivo Parquet
            df = pd.read_parquet(os.path.join(parquet_path, filename))
            
            # Converter DataFrame para dicionário para inserção no Elasticsearch
            records = df.to_dict(orient='records')
            
            # Preparar os dados para inserção usando a Bulk API
            actions = [
                {
                    "_index": "teste",  # substitua pelo nome do seu índice
                    "_source": record
                }
                for record in records
            ]
            
            # Inserir os dados no Elasticsearch
            helpers.bulk(es, actions)

# Executar a função
index_parquet_files()
```

```python
# Define o índice a ser pesquisado
index = "teste"

# Define a consulta de busca
query = {
    "query": {
        "match_all": {}  # Esta é uma consulta simples que retorna todos os documentos
    },
    "size": 200
}
```

```python
# Executar a busca
response = es.search(index=index, body=query)

# Imprimir os resultados
for hit in response['hits']['hits']:
    print(hit['_source'])
```
