```python
import pandas as pd
```

| Campo                          | Descrição                                                  |
| ------------------------------ | ---------------------------------------------------------- |
| Transaction unique identifier  | Identificador único da transação                           |
| Price                          | Preço da propriedade                                       |
| Date of Transfer               | Data da transferência da propriedade                       |
| Property Type                  | Tipo de propriedade (por exemplo, casa, apartamento, etc.) |
| Old/New                        | Indica se a propriedade é nova ou antiga                   |
| Duration                       | Duração do contrato de propriedade (por exemplo, arrendamento, propriedade total) |
| Town/City                      | Cidade ou município onde a propriedade está localizada     |
| District                       | Distrito onde a propriedade está localizada                |
| County                         | Condado onde a propriedade está localizada                 |
| PPDCategory Type               | Tipo de categoria de dados da PPD (Public Price Data)      |
| Record Status - monthly file only | Status do registro (apenas em arquivos mensais)          |
